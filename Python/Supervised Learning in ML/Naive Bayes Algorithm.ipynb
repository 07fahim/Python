```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
dataset=pd.read_csv("D:/Udemy-GenAI/placement (1).csv")
dataset.drop(columns=['Unnamed: 0'],inplace=True)
dataset.head()
```

```python
sns.pairplot(dataset,hue='placement')
plt.show()
```

```python
x=dataset.iloc[:,:-1]
y=dataset['placement']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
from sklearn.naive_bayes import GaussianNB,BernoulliNB,MultinomialNB

gnb=GaussianNB()
gnb.fit(x_train,y_train)
```

```python
gnb.score(x_test,y_test)*100,gnb.score(x_train,y_train)*100
```

```python
from mlxtend.plotting import plot_decision_regions

plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=gnb)
plt.show()
```

```python
bnb=BernoulliNB()
bnb.fit(x_train,y_train)
```

```python
bnb.score(x_test,y_test)*100,bnb.score(x_train,y_train)*100
```

```python
from mlxtend.plotting import plot_decision_regions

plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=bnb)
plt.show()
```
