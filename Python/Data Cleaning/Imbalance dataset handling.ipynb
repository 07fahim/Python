```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
dataset=pd.read_csv("D:/Udemy-GenAI/Social_Network_Ads.csv")
dataset.drop(columns=['EstimatedSalary'],inplace=True)
dataset.head()
```

```python
sns.scatterplot(x=dataset['Age'],y=dataset['Purchased'],data=dataset)
plt.show()
```

```python
x=dataset[['Age']]
y=dataset['Purchased']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression

lr=LogisticRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
lr.predict([[19]])
```

```python
# Scatter plot for original data
sns.scatterplot(x=dataset['Age'], y=dataset['Purchased'], label='Original Data')

# Line plot for predicted values
sns.lineplot(x=dataset['Age'], y=lr.predict(x), color='red', label='Prediction Line')

# Add legend, title, and labels
plt.legend()
plt.title("Linear Regression Prediction")
plt.xlabel("Age")
plt.ylabel("Purchased")
plt.show()
```
