## 4.Distplot Nu-Cat


```python
sns.distplot(titanic[titanic['Survived'] == 0]['Age'],hist=False,color='red',label='Dead')
sns.distplot(titanic[titanic['Survived'] == 1]['Age'],hist=False,color='green',label='Alive')
plt.legend()
plt.show()
```

```python
sns.distplot(hotel[hotel['is_canceled'] == 0]['adults'],hist=False,color='red',label='Dead')
sns.distplot(hotel[hotel['is_canceled'] == 1]['adults'],hist=False,color='green',label='Alive')
plt.legend()
plt.show()
```
