```python
import numpy  as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
titanic=pd.read_csv("D:/Udemy-GenAI/Titanic-Dataset.csv")
titanic.head()
```

```python
tips=pd.read_csv("D:/Udemy-GenAI/tips.csv")
tips.head()
```

```python
iris=pd.read_csv("D:/Udemy-GenAI/Iris.csv")
iris.head()
```

```python
hotel=pd.read_csv("D:/Udemy-GenAI/hotel_booking.csv/hotel_booking.csv")
hotel.head()
```

## 1.Scatterplot-Nu-nu

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x=tips['total_bill'],y=tips['tip'],hue=tips['sex'],style=tips['smoker'],size=tips['size'])
```

## 2.Barplot Nu-Cat

```python
sns.barplot(x=titanic["Pclass"],y=titanic['Age'],hue=titanic['Sex'])
```

```python
sns.barplot(x=titanic["Pclass"],y=titanic['Fare'],hue=titanic['Sex'])
```

## 3.Boxplot Nu-Cat

```python
plt.figure(figsize=(10,6))
sns.boxenplot(x=titanic["Pclass"],y=titanic["Age"],hue=titanic['Sex'])
```

```python
plt.figure(figsize=(10,6))
sns.boxenplot(x=titanic["Sex"],y=titanic["Age"],hue=titanic['Survived'])
```

## 4.Distplot Nu-Cat


```python
sns.distplot(titanic[titanic['Survived'] == 0]['Age'],hist=False,color='red',label='Dead')
sns.distplot(titanic[titanic['Survived'] == 1]['Age'],hist=False,color='green',label='Alive')
plt.legend()
plt.show()
```

```python
sns.distplot(hotel[hotel['is_canceled'] == 0]['adults'],hist=False,color='red',label='Dead')
sns.distplot(hotel[hotel['is_canceled'] == 1]['adults'],hist=False,color='green',label='Alive')
plt.legend()
plt.show()
```

## 5.HeatMap NU-NU

```python
sns.heatmap(pd.crosstab(titanic['Pclass'],titanic['Survived']))
```

## 6.ClusterPlot

```python
sns.clustermap(pd.crosstab(titanic['Pclass'],titanic['Survived']))
```
