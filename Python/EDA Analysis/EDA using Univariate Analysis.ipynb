## 1.Categorical Data

a.Countplot

```python
sns.countplot(x=dataset['Survived'],data=dataset,color='red')
dataset['Survived'].value_counts()
```

```python
sns.countplot(x=dataset['Sex'],data=dataset,color='blue')
dataset['Sex'].value_counts()
```

```python
sns.countplot(x=dataset['Pclass'],data=dataset,color='green')
dataset['Pclass'].value_counts()
```

b.pie chart

```python
dataset['Embarked'].value_counts().plot(kind='pie', autopct='%.2f', title='Distribution of Embarked',shadow=True,explode=(0.0001, 0.1, 0.1))
```

```python
dataset['Sex'].value_counts().plot(kind='pie', autopct='%.2f', title='Distribution of Gender',shadow=True,explode=(0.001, 0.1))
```
